```python
#Calculate percent weight difference
def pcwc(date1, date2):
    return 100*(weights[date2] - weights[date1])/weights[date1]
```

```python
#Add weight differences to dataframe
def apply_wc(recent, now, dataset):
    wFI = {'mouse': weights['NAME'], '%twc': pcwc('9/20/17', now), '%rwc': pcwc(recent, now)}
    wFI = pd.DataFrame(data = wFI)
    wFI.set_index('mouse', inplace=True)
    FI_wFI = pd.merge(dataset, wFI, left_index=True, right_index=True)
    return FI_wFI
```

```python
#Make the first line the header in a dataframe
def firstline_header(x):
    new_header = x.iloc[0] #grab the first row for the header
    x = x[1:] #take the data less the header row
    x.columns = new_header #set the header row as the df header
    return x
```

```python
#Import and setup enalapril data
def enalapril_data_setup(data):
    data = firstline_header(data)
    data = data.rename(columns=lambda x: x.replace(' ', '_'))
    data = data.rename(columns={'Eye_discharge/swelling': 'Eye_discharge_swelling', 'Vaginal/uterine/penile_prolapse': 'Vaginal_uterine_penile_prolapse', 'Breathing_rate/depth': 'Breathing_rate_depth'})
    #Add threshold %rwc (=1 if <10% or >8% )
    data['Thresh_%rwc'] = ((data['%rwc'] < -10) | (data['%rwc'] > 8))*1
    #Fill missing data with mean (only %rwc and %twc have missing data)
    data['%rwc'] = data['%rwc'].transform(lambda x: x.fillna(x.mean()))
    data['%twc'] = data['%twc'].transform(lambda x: x.fillna(x.mean()))
    return(data)
```

```python
#Calculate percent recent weight loss for Mitchell data
def Mitchell_perc_rwc(n):
    bwnow = Mitchell_all.iloc[n]['bwAbs']
    time = Mitchell_all.iloc[n]['Time']
    mouse= Mitchell_all.index[n]
    bwbefore = Mitchell_all[(Mitchell_all.index==mouse) & (Mitchell_all['Time']==time-1)]['bwAbs']
    if len(bwbefore)==0:
        return 0
    else:
        return (100*(bwnow-bwbefore)/bwbefore)[0]
```

## Statistics

```python
#Print statistics for a correlation
def print_stats(x,y, dec=3):
    slope, intercept, r_value, p_value, std_err = sp.stats.linregress(x,y)
    
    Mr = abs(x-y).mean()
    Medr = np.median(abs(x-y))
    Stdr = np.std(x-y)
    Mpr = np.median(100*abs((x-y)/x))
    rs = r_value**2
    
    print("Mean residual  =", round(Mr, dec))
    print("Median residual  =", round(Medr, dec))
    print("Stdev residual =", round(Stdr, dec))
    print("Median percent residual =", round(Mpr, dec),'%')
    print("r_value =", round(r_value, dec))
    print("r_squared =", round(rs, dec))
    print("p_value =", p_value)
    print("std-err =", round(std_err, dec))  
```

```python
#Print r-squared and p-value only
def print_short_stats(x,y, dec=3):
    slope, intercept, r_value, p_value, std_err = sp.stats.linregress(x, y)
    
    Mr = abs(x-y).mean()
    Medr = np.median(abs(x-y))
    Stdr = np.std(x-y)
    Mpr = np.median(100*abs((x-y)/x))
    rs = r_value**2
    
    print("r_squared =", round(rs, dec))
    print("p_value =", p_value)
```

```python
#Return statistics for a correlation
def return_stats(x,y, dec=3):
    slope, intercept, r_value, p_value, std_err = sp.stats.linregress(x,y)
    
    Mr = abs(x-y).mean()
    Medr = np.median(abs(x-y))
    Mpr = np.median(100*abs((x-y)/x))
    rs = r_value**2
    
    return(Mr, Medr, Mpr, p_value, rs)
```

```python
#Return r-squared and p-value only
def return_short_stats(x,y, dec=3):
    slope, intercept, r_value, p_value, std_err = sp.stats.linregress(x, y)
    
    Mr = abs(x-y).mean()
    Medr = np.median(abs(x-y))
    Stdr = np.std(x-y)
    Mpr = np.median(100*abs((x-y)/x))
    rs = r_value**2
    
    return [round(rs, dec), p_value]
```

```python
#For AFRAID scores, at a given age(s), identify the top and bottom predicted quartiles and print how long they actually lived
def print_quartiles(ages):
    dataset = daysalivetest[daysalivetest['months'].isin(ages)]
    bottom_cutoff = np.percentile(dataset['pred'], 25)
    top_cutoff = np.percentile(dataset['pred'], 75)
    bottom = dataset[dataset['pred'] <= bottom_cutoff]['Days_alive']
    top = dataset[dataset['pred'] >= top_cutoff]['Days_alive']
    print('Bottom Quartile')
    print(list(round(bottom,2)))
    print('Mean: ', bottom.mean())
    print('Top Quartile')
    print(list(round(top,2)))
    print('Mean: ', top.mean())
    print('P-value: ', round(stats.ttest_ind(top, bottom).pvalue,4))
```

```python
#Wiggle datapoints for a graph for easier viewing (for when ages are the same)
def wiggle(x, size=100):
    rands = list([(random.randint(-size, size)/100) for k in range(len(x)) ])
    x_wiggle = list( map(add, x, rands))
    return x_wiggle
```

```python
#Plot a graph for prediction of age
def plot_fright_graph(x, y, title):
    fig = plt.figure(figsize=(6,6))
    plt.style.use('classic')
    plt.xlim([20,40])
    plt.xlabel('Predicted age (months)', fontsize=18)
    plt.ylim([20,40])
    plt.ylabel('True age (months)', fontsize=18)
    plt.xticks(np.arange(21,42,3), fontsize=14)
    plt.yticks(np.arange(21,42,3), fontsize=14)
    plt.title(title, fontsize=18)
    sns.regplot(x, wiggle(y), scatter_kws={'s':50}, fit_reg=True, ci=0)
    a=[20,30,40]
    b=[20,30,40]
    plt.plot(a,b,'k--')
    fig.patch.set_facecolor('white')
    plt.show()
    print(title + ' stats:')
    print('')
    print_stats(x, y)
```

```python
def plot_delta_age_by_age(dataset):
    months = np.sort(dataset['age_months'].unique())
    sns.set(rc={'figure.figsize':(6,6)})
    sns.set_style('white')
    rs_thresh = 0.1
    colors = sns.color_palette("hls", len(months))
    handle_list = []
    n = 0
    for month in months:
        data = dataset[dataset['age_months']==month]
        x = data['delta age']
        y = data['Months_alive']
        if (return_short_stats(x,y)[0] >= rs_thresh):
            sns.regplot(x=x, y=y, color=colors[n], fit_reg=True, ci=0, truncate=True, scatter_kws={"s": 50})
        else:
            sns.regplot(x=x, y=y, color=colors[n], fit_reg=False, scatter_kws={"s": 50})
        handle_now = mlines.Line2D([], [], color=colors[n], marker='.' , markersize=15, linestyle='None', label=month)
        handle_list.append(handle_now)
        n=n+1

    plt.xlabel('Delta age (months)', fontsize=18)
    plt.ylabel('Survival (months)', fontsize=18)
    plt.xlim([-8,8])
    plt.ylim([0,20])
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    fig.patch.set_facecolor('white')
    plt.legend(handles=handle_list, numpoints=1, loc='upper left', title='Age')
    plt.show()
```

## KDM

```python
def KDM_model(trainset, testset, cachename):
    #print(cachename)
    trainset['seqn'] = range(1,trainset.shape[0]+1)
    trainset['group'] = 1
    trainset['samp_wt'] = 1
    
    testset['seqn'] = range(1,testset.shape[0]+1)
    testset['group'] = 1
    testset['samp_wt'] = 1
    
    testset = testset.drop(['GROUP', 'Days_alive', 'age_months'], axis = 1)
    trainset = trainset.drop(['GROUP', 'Days_alive', 'age_months'], axis = 1)
    #print(testset.shape, trainset.shape)
    
    # change format of train set to old format that BA method requires
    header_names = list(trainset)
    trainset_list = trainset.values.tolist()
    trainset_list.insert(0, header_names)
    
    testset_list = testset.values.tolist()
    testset_list.insert(0, header_names)
    
    # set indices for age, group, primary index, sample weights (just a column of ones)
    age_index = header_names.index('age_days')
    genderindex = header_names.index('group')
    primaryindex = header_names.index('seqn')
    samp_wt_index = header_names.index('samp_wt')
    
    # Build model
    model = BA.Methods(trainset_list, cachename, age = age_index, genderindex = genderindex, primarykey = primaryindex, samp_wt = samp_wt_index)
    model.KDM() # train model
    
    # change format of test set to old format that BA method requires
    header_names = list(testset)
    testset_list = testset.values.tolist()
    testset_list.insert(0, header_names)
    
    # test
    test_model = BA.Methods(testset_list, cachename, age = age_index, genderindex = genderindex, primarykey = primaryindex, samp_wt = samp_wt_index)
    results = test_model.KDM()
    results = pd.DataFrame(results[1:], columns = results[0])
    # test_results = test_results.append(results) # this was designed for 5-fold cross validation, so it would append the results of each fold into one dataframe

    # calculate stats
    stats = return_stats(results['BA'].astype('float64'), results['age_days'].astype('float64'))
    
    return (stats)
```

## Methionine restriction

### Data import and setup

```python
#Data import and setup
Mitchell_all = pd.read_csv('MitchellDatasetAll.csv') #Import data
Mitchell_all = Mitchell_all.set_index(['Mouse']) #Set mouse as index
Mitchell_all = Mitchell_all[Mitchell_all['age_months'] >= 20] #Only consider mice older than 20
Mitchell_all['age_months'] =  round(Mitchell_all['age_months']) #Round  to the nearest month
```

```python
#Get baseline weight for each mouse
baseline_weights = pd.DataFrame(columns = ['baseline_weight'])
Mitchell_mice = list(Mitchell_all.index.unique())
for mouse in Mitchell_mice:
    blweight = list(Mitchell_all[Mitchell_all.index == mouse]['bwAbs'])[0]
    baseline_weights.loc[mouse] = blweight
#Mirror the order of mice in the Mitchall_all dataframe
baseline_weights_mirror = []
for mouse in Mitchell_all.index:
    baseline_weights_mirror.append(baseline_weights.loc[mouse][0])
#Calculate and add %twc
Mitchell_all['%twc'] = 100*(Mitchell_all['bwAbs'] - baseline_weights_mirror)/baseline_weights_mirror
#Calculate and add %rwc
Mitchell_perc_rwc_list = []
for n in range(0,len(Mitchell_all.index)):
    Mitchell_perc_rwc_list.append(Mitchell_perc_rwc(n))
Mitchell_all['%rwc'] = Mitchell_perc_rwc_list
#Calculate and add Thresh_%rwc (=1 if <10% or >8% )
Mitchell_all['Thresh_%rwc'] = ((Mitchell_all['%rwc'] < -10) | (Mitchell_all['%rwc'] > 8))*1
```

```python
#Get control and MetR data at 27 months
Mitchell_control_27 = Mitchell_all[(Mitchell_all['Group']=='Con')&(Mitchell_all['age_months']==27)]
Mitchell_MetR_27 = Mitchell_all[(Mitchell_all['Group']=='MetR')&(Mitchell_all['age_months']==27)]
```
